```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['cityCode_encoded'] = label_encoder.fit_transform(df['cityCode'])
print(df)
```

```python
df1 = df.drop('cityCode', axis=1)
```

```python
df1.head()
```

## Linear Regression Model

```python
X = df1[['squareMeters', 'numberOfRooms', 'hasYard', 'hasPool', 'floors', 'cityPartRange', 'numPrevOwners', 'made', 'isNewBuilt', 'hasStormProtector', 'basement', 'attic', 'garage', 'hasStorageRoom', 'hasGuestRoom', 'cityCode_encoded']]
y = df1['price']
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,random_state=42,test_size=0.32)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
round_mse = round(mse,2)
print('Mean Squared Error Linear Regression Model:', round_mse)
```

```python
new_data = pd.DataFrame({
    'squareMeters': [60000],
    'numberOfRooms': [4],
    'hasYard': [1],
    'hasPool': [1],
    'floors': [2],
    'cityPartRange': [5],
    'numPrevOwners': [4],
    'made': [2010],
    'isNewBuilt': [1],
    'hasStormProtector': [0],
    'basement': [1],
    'attic': [1],
    'garage': [2],
    'hasStorageRoom': [1],
    'hasGuestRoom': [0],
    'cityCode_encoded': [1234]
})
predicted_price = model.predict(new_data)
print('Predicted Price:', predicted_price)
```

## Decision-Tree and Random Forest Model

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
## DECISION TREE MODEL

model_dt = DecisionTreeRegressor()
model_dt.fit(X_train , y_train)
```

```python
## RANDOM FOREST MODEL

model_rf = RandomForestRegressor()
model_rf.fit(X_train , y_train)
```

```python
## Mean Squared Error Decision Tree

y_pred_dt = model_dt.predict(X_test)
mse2 = mean_squared_error(y_test, y_pred_dt)
round_mse1 = round(mse2,2)
print('Mean Squared Error Decision Tree Model:', round_mse1)
```

```python
## Mean Squared Error Random Forest

y_pred_rf = model_rf.predict(X_test)
mse3 = mean_squared_error(y_test, y_pred_rf)
round_mse2 = round(mse3,2)
print('Mean Squared Error Random Forest Model:', round_mse2)
```
